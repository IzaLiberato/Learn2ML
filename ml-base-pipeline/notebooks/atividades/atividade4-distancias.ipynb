# 2
Transforme os dados contínuos de idade em dados categóricos de acordo com uma regra. Sugestão de regra: 

0 a 18 anos: criança

18 a 60 anos: adulto

maior que 60: idoso

```python
df = df_original.copy()
df = df.dropna()

#df
```

```python
ages = []

for i in df['Age']:
    if i > 0 and i <=18:
        ages.append('criança')
    if i >18 and i<=60:
        ages.append('adulto')
    if i > 60:
         ages.append('idoso')

df['Age'] = ages
df
```

Crie 3 vetores de atributos distintos, e utilizando as distâncias euclidiana e Manhattan, determine a qual classe cada vetor de atributos pertence.

```python
from scipy.spatial.distance import cityblock
from math import dist

point_1 = (1,2,4)
point_2 = (1,1,1)

point = np.array(point_1) 
point2 = np.array(point_2)
#point3 = np.array(point3)
sum_sq = np.sum(np.square(point)) #EUCLIDIANA (city block)
cityblock_c = cityblock(point_1, point2)

print('Distância Euclidiana:',np.sqrt(sum_sq))

if sum_sq == 10:
    print('classe 2')

if sum_sq <= 10:
    print('classe 2')

if sum_sq >= 11:
    print('classe 3')
```

```python
def manhattan_distance(point1, point2):
    distance = 0
    for x1, x2 in zip(point1, point2):
        difference = x2 - x1
        absolute_difference = abs(difference)
        distance += absolute_difference

    return distance

point_1 = (1,2,4)
point_2 = (1,1,1)
md = manhattan_distance(point_1, point_2)
print('Distância Manhattan:',md)

if md == 10:
    print('classe 2')

if md <= 10:
    print('classe 2')

if md >= 11:
    print('classe 3')
```
